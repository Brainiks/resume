```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df= pd.read_csv('/content/preobraz_train_new.csv')

# Преобразование столбца с датами
df['event_timestamp'] = pd.to_datetime(df['event_timestamp'], errors='coerce')

# Проверка преобразования
print(df['event_timestamp'].head())

# Удаление строк с пустыми значениями во всех столбцах
df= df.dropna()

# Сохранение очищенного датасета


# Подсчёт количества мужчин и женщин в каждом городе
sex_counts_per_city = df.groupby(['region', 'sex']).size().unstack(fill_value=0)

# Проверяем, какие столбцы создались
print(sex_counts_per_city)

# Подсчёт вероятности быть женщиной в каждом городе
sex_counts_per_city['prob_woman'] = sex_counts_per_city['female'] / (sex_counts_per_city['male'] + sex_counts_per_city['female'])

# Присоединение вероятности к основному датасету
df = df.merge(sex_counts_per_city[['prob_woman']], on='region', how='left')

# Просмотр обновлённого датафрейма


# Группировка данных по категориям и полу
category_sex_counts = df.groupby(['category', 'sex']).size().unstack(fill_value=0)

# Рассчёт вероятности быть женщиной в каждой категории
category_sex_counts['prob_woman'] = category_sex_counts['female'] / (category_sex_counts['female'] + category_sex_counts['male'])

# Соединение с основным датафреймом, чтобы добавить новый столбец
df = df.merge(category_sex_counts[['prob_woman']], on='category', how='left')

# Переименование нового столбца для ясности
df.rename(columns={'prob_woman': 'prob_woman_categories'}, inplace=True)

# Просмотр обновлённого датафрейма




# Подсчет общего количества пользователей по городам и возрастным категориям
age_counts_cities = df.groupby(['region', 'age_class']).size().unstack(fill_value=0)

# Подсчет вероятности принадлежности к каждой возрастной категории по городам
probabilities_cities = age_counts_cities.div(age_counts_cities.sum(axis=1), axis=0)

# Добавление вероятностей в оригинальный датафрейм
for age_class in probabilities_cities.columns:
    df[f'prob_{age_class}_cities'] = df['region'].map(probabilities_cities[age_class])

# Проверка результата
print(df[['region', 'age_class'] + [f'prob_{age_class}_cities' for age_class in probabilities_cities.columns]].head())



# Подсчет общего количества пользователей по категориям видео и возрастным категориям
age_counts_categories = df.groupby(['category', 'age_class']).size().unstack(fill_value=0)

# Подсчет вероятности принадлежности к каждой возрастной категории по категориям видео
probabilities_categories = age_counts_categories.div(age_counts_categories.sum(axis=1), axis=0)

# Добавление вероятностей в оригинальный датафрейм
for age_class in probabilities_categories.columns:
    df[f'prob_{age_class}_categories'] = df['category'].map(probabilities_categories[age_class])

# Проверка результата
print(df[['category', 'age_class'] + [f'prob_{age_class}_categories' for age_class in probabilities_categories.columns]].head())

# Группировка по ua_device_type и полу
sex_counts_device = df.groupby(['ua_device_type', 'sex']).size().unstack(fill_value=0)
# Подсчет вероятности, что пользователь — женщина
prob_female_device = sex_counts_device['female'] / (sex_counts_device['female'] + sex_counts_device['male'])

# Группировка по ua_client_type и полу
sex_counts_client = df.groupby(['ua_client_type', 'sex']).size().unstack(fill_value=0)
# Подсчет вероятности, что пользователь — женщина
prob_female_client = sex_counts_client['female'] / (sex_counts_client['female'] + sex_counts_client['male'])

# Группировка по ua_os и полу
sex_counts_os = df.groupby(['ua_os', 'sex']).size().unstack(fill_value=0)
# Подсчет вероятности, что пользователь — женщина
prob_female_os = sex_counts_os['female'] / (sex_counts_os['female'] + sex_counts_os['male'])

# Группировка по ua_client_name и полу
sex_counts_client_name = df.groupby(['ua_client_name', 'sex']).size().unstack(fill_value=0)
# Подсчет вероятности, что пользователь — женщина
prob_female_client_name = sex_counts_client_name['female'] / (sex_counts_client_name['female'] + sex_counts_client_name['male'])

# Добавляем в датафрейм вероятности
df['prob_female_device'] = df['ua_device_type'].map(prob_female_device)
df['prob_female_client'] = df['ua_client_type'].map(prob_female_client)
df['prob_female_os'] = df['ua_os'].map(prob_female_os)
df['prob_female_client_name'] = df['ua_client_name'].map(prob_female_client_name)

# Проверка результата
print(df[['ua_device_type', 'ua_client_type', 'ua_os', 'ua_client_name',
          'prob_female_device', 'prob_female_client', 'prob_female_os', 'prob_female_client_name']].head())


# Список возрастных категорий
age_classes = df['age_class'].unique()

# Группировка по ua_device_type и возрастной категории
age_counts_device = df.groupby(['ua_device_type', 'age_class']).size().unstack(fill_value=0)
# Расчет вероятности для каждой возрастной группы по ua_device_type
prob_age_device = age_counts_device.div(age_counts_device.sum(axis=1), axis=0)

# Группировка по ua_client_type и возрастной категории
age_counts_client = df.groupby(['ua_client_type', 'age_class']).size().unstack(fill_value=0)
# Расчет вероятности для каждой возрастной группы по ua_client_type
prob_age_client = age_counts_client.div(age_counts_client.sum(axis=1), axis=0)

# Группировка по ua_os и возрастной категории
age_counts_os = df.groupby(['ua_os', 'age_class']).size().unstack(fill_value=0)
# Расчет вероятности для каждой возрастной группы по ua_os
prob_age_os = age_counts_os.div(age_counts_os.sum(axis=1), axis=0)

# Группировка по ua_client_name и возрастной категории
age_counts_client_name = df.groupby(['ua_client_name', 'age_class']).size().unstack(fill_value=0)
# Расчет вероятности для каждой возрастной группы по ua_client_name
prob_age_client_name = age_counts_client_name.div(age_counts_client_name.sum(axis=1), axis=0)

# Добавляем в датафрейм вероятности для каждой возрастной группы
for age_class in age_classes:
    df[f'prob_{age_class}_device'] = df['ua_device_type'].map(prob_age_device[age_class])
    df[f'prob_{age_class}_client'] = df['ua_client_type'].map(prob_age_client[age_class])
    df[f'prob_{age_class}_os'] = df['ua_os'].map(prob_age_os[age_class])
    df[f'prob_{age_class}_client_name'] = df['ua_client_name'].map(prob_age_client_name[age_class])

# Проверка результата
print(df[['ua_device_type', 'ua_client_type', 'ua_os', 'ua_client_name'] +
         [f'prob_{age_class}_device' for age_class in age_classes] +
         [f'prob_{age_class}_client' for age_class in age_classes] +
         [f'prob_{age_class}_os' for age_class in age_classes] +
         [f'prob_{age_class}_client_name' for age_class in age_classes]].head())


df.to_csv('/content/test.csv', index=False)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df= pd.read_csv('/content/preobraz_train_new.csv')

# Предполагается, что ваш DataFrame называется df
# И загружен он из CSV файла, который содержит столбец 'titles'

# Создание векторизатора
# Ограничение количества фичей, например, до 5000
vectorizer = TfidfVectorizer(max_features=100)
tfidf_matrix = vectorizer.fit_transform(df['title'])


# Получение векторов в виде массива
tfidf_array = tfidf_matrix.toarray()

# Получение названий фичей (слов)
feature_names = vectorizer.get_feature_names_out()

# Создание DataFrame для удобства работы
tfidf_df = pd.DataFrame(tfidf_array, columns=feature_names)




# Сохранение в CSV файл
tfidf_df.to_csv('tfidf_video_titles.csv', index=False)

# Печать первых 5 строк результирующего DataFrame
print(tfidf_df.head())
```
